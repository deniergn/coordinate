```python
import math
import numpy as np
import csv

def circle_points(r, n):  #r: radius of the circle, n: number of points on the circle
    
    coordinates = []
    
    for i in range(n):
        theta = 2*(i/n)*math.pi
        
        x = r*np.round(np.cos(theta), decimals=5) # w/o rounding it gives awkward results, since pi is a float
        y = r*np.round(np.sin(theta), decimals=5)
        
        coordinates.append((x,y))
        
        
# Creating the csv file

    with open('coordinate.csv', 'w', encoding='UTF8') as file:  
        writer = csv.writer(file)
        writer.writerow(['x','y'])
        
        for row in coordinates:  #stores the coordinates in seperate rows
            writer.writerow(row)
        
        # writer.writerow(coordinates) stores the coordinates as tuples
        
    return file

circle_points(1,1000)
    
```
